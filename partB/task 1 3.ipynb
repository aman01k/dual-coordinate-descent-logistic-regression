**Paper:** *Dual Coordinate Descent Methods for Logistic Regression and Maximum Entropy Models*  
**Authors:** Yu, Huang, Lin (2011) — *Machine Learning*, 85:41–75

---

## Baseline Method Identified

The primary baselines the paper compares against are:
1. **TRON** — Trust Region Newton method (Lin, Weng & Keerthi, 2008) for logistic regression.
2. **CDprimal** — Primal coordinate descent for logistic regression (Huang et al., 2010).
3. **L-BFGS** — Limited-memory quasi-Newton method (Liu & Nocedal, 1989).

The **most directly compared and primary baseline is CDprimal** (primal coordinate descent by Huang et al., 2010), because the paper's method (CDdual) is explicitly positioned as a *dual* counterpart of CDprimal, and it is the baseline used to motivate the dual approach throughout Section 2.2 and the experiments.

---

## Limitation of CDprimal Identified by the Paper

CDprimal (Section 2.2) applies coordinate descent to the primal LR objective by updating one weight $w_j$ at a time. Each one-variable sub-problem in the primal (Eq. 12) does **not** have a closed-form solution: finding the Newton direction requires computing $g'(z)$ and $g''(z)$ (Eqs. 13–14), each of which involves $l$ expensive exponential evaluations ($O(l)$ cost). Additionally, each line-search step requires further exponential evaluations to check sufficient decrease. Because these operations are expensive, Huang et al. apply only *one* Newton iteration per variable — a deliberate approximation to limit per-step cost. This means CDprimal must iterate through all $n$ primal variables per pass with a high per-iteration constant, making it slower than CDdual for datasets where $l \gg n$ (more instances than features).

---

## How CDdual Overcomes This Limitation

By working in the dual, the paper reduces the per-iteration cost of constructing the sub-problem to $O(n)$ (one dot product against the maintained $w(\alpha)$). The Newton direction and all subsequent line-search-equivalent steps in Algorithm 4 cost only $O(1)$ (no exponential operations, only log evaluations which are handled numerically via reformulation). Because each step is cheap, the algorithm can *accurately* solve the sub-problem (multiple Newton iterations per variable) rather than approximating it, yielding better per-pass progress. The result (Table 1, Section 3.1) is that CDdual's per-round cost is $O(l \times \text{\#Newton steps})$ vs. CDprimal's $O(nl)$, making CDdual at least 6× faster per round on datasets where $n$ is large.

---

## Condition Under Which CDdual Would NOT Outperform CDprimal

The paper itself notes (Section 6.1, discussion of **a9a** dataset) that CDdual is *not* consistently faster than CDprimal — or TRON/L-BFGS — when the number of features $n$ is much smaller than the number of training instances $l$. In such a regime, the primal problem has fewer variables ($n$) than the dual ($l$), so a single pass through the primal is cheaper than a single pass through the dual. The a9a dataset has $l = 32{,}561$ instances but only $n = 123$ features; on this dataset CDdual is actually slower than CDprimal and TRON (Fig. 2, first row). More generally, CDdual would not outperform primal methods whenever the dataset is low-dimensional and high-sample — a common situation in classical tabular datasets such as census or credit-scoring data — because the $O(n)$ per-step advantage of the dual disappears relative to the $O(n)$ cost of the primal when $n \ll l$.
